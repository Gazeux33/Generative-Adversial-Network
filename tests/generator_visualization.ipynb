```python
nb_images_per_row = 10
nb_images = 100
```

```python
with torch.no_grad():
    images = generator(torch.randn(nb_images,config.Z_DIM))
```

```python
nrow = nb_images//nb_images_per_row
nrow
```

```python

fig, ax = plt.subplots(nrows=nb_images_per_row,ncols=nb_images_per_row, figsize=(10, 10))
plt.subplots_adjust(wspace=0, hspace=0)


for i in range(nrow):
    for j in range(nb_images_per_row):
        img = images[i*nb_images_per_row+j].detach().cpu().numpy()
        img = np.transpose(img, (1, 2, 0))
        ax[i,j].imshow(img)
        ax[i,j].axis('off')
        
```
